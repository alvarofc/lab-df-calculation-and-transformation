```python
items = stores2.drop(columns=['date','Unnamed: 0','shop_id'])
```

```python
items['Total'] = items['item_price']*items['item_cnt_day']
```

```python
items_total = items.groupby('item_id').sum()
items_total.head()
```

```python
# Best selling items by number of sold items
items_total.nlargest(5,'item_cnt_day')
```

```python
# Best selling items by revenue
items_total.nlargest(5,'Total')
```

```python
#Lowest selling items with lowest earnings
items_total.sort_values(['Total','item_cnt_day']).nsmallest(5,'item_cnt_day')
```
